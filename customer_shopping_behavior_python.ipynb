```python
import pandas as pd

# Load CSV
df = pd.read_csv("Customer_Shopping_Behavior.csv")

# Clean column names for SQL (remove spaces, parentheses, hyphens)
df.columns = (
    df.columns
    .str.strip()
    .str.replace(' ', '_', regex=False)
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
    .str.replace('-', '_', regex=False)
)

# Replace NaN with None so SQL Server can accept them as NULL
df = df.where(pd.notnull(df), None)
```

```python
# Drop table if it exists
cursor.execute("IF OBJECT_ID('dbo.Customer', 'U') IS NOT NULL DROP TABLE dbo.Customer")
conn.commit()

# Map pandas dtypes to SQL Server types
dtype_map = {
    'int64': 'INT',
    'float64': 'FLOAT',
    'object': 'NVARCHAR(255)',
    'datetime64[ns]': 'DATE'
}

columns_with_types = []
for col in df.columns:
    col_type = dtype_map.get(str(df[col].dtype), 'NVARCHAR(255)')
    columns_with_types.append(f"[{col}] {col_type}")

# Create table
create_table_sql = f"CREATE TABLE dbo.Customer ({', '.join(columns_with_types)})"
cursor.execute(create_table_sql)
conn.commit()
print("Table created successfully!")
```

```python
import numpy as np

for index, row in df.iterrows():
    placeholders = ", ".join(["?"] * len(df.columns))
    insert_sql = f"INSERT INTO dbo.Customer ({', '.join(df.columns)}) VALUES ({placeholders})"

    # Convert each value to Python-native types
    row_values = []
    for col in df.columns:
        val = row[col]
        if pd.isna(val):
            val = None
        elif isinstance(val, (np.integer, np.int64)):
            val = int(val)
        elif isinstance(val, (np.floating, np.float64)):
            val = float(val)
        elif isinstance(val, pd.Timestamp):
            val = val.to_pydatetime()
        row_values.append(val)

    cursor.execute(insert_sql, row_values)

conn.commit()
print("Data inserted successfully!")
```
